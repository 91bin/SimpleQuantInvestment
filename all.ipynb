```python
# key(재무제표 항목들) 확인 코드
import json
with open('./data/ticker_cik_mapping.json', 'r', encoding='utf-8') as f:
    t_to_c = json.load(f)
base_path = "./data/companyfacts/" 

ticker = 'AAPL'  # 여기에 원하는 ticker 이름을 입력하면 됩니다.

cik_file_name ="CIK"+t_to_c[ticker.upper()]+".json"
filepath = base_path + cik_file_name
with open(filepath, 'r') as f:
    json_data = json.load(f)

print(json_data['facts'].keys())
with open(f"./data/keys_{ticker}.txt", 'w') as f:
    f.write(str(json_data['facts']['us-gaap'].keys()))

# 결과물로 생긴 keys_ticker 파일 확인
```
